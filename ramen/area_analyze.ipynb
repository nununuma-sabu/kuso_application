# 必要なライブラリのインポート
ライブラリをインポートします。

```python
# 必要なライブラリのインポート
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

# CSVファイルの読み込み
CSVファイルを読み込みます。

```python
# CSVファイルの読み込み
df = pd.read_csv('grouped_middle_area.csv')
```

# 大エリアのエリアごとの件数表示

```python
# 大エリアごとの件数を表示
area_counts = df.groupby('大エリア')['件数'].sum().sort_values(ascending=False)
print(area_counts)
```

# 棒グラフでプロット

```python
# 棒グラフでプロット
plt.figure(figsize=(10, 6))
area_counts.plot(kind='bar')
plt.title('大エリアごとの件数')
plt.xlabel('大エリア')
plt.ylabel('件数')
plt.show()
```

```python
total = area_counts.sum()
print('全体の件数:', total)
```

# 中エリアのエリアごとの件数

```python
# 大エリアごとの件数を表示
area_counts2 = df.groupby('中エリア')['件数'].sum().sort_values(ascending=False)
print(area_counts2)

# 棒グラフでプロット
plt.figure(figsize=(10, 6))
area_counts2.plot(kind='bar')
plt.title('中エリアごとの件数')
plt.xlabel('中エリア')
plt.ylabel('件数')
plt.show()
```
